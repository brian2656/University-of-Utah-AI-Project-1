```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
```

```python
path_population = Path("Resources/population(1).csv")
population_df = pd.read_csv(path_population)
```

```python
selected_columns = population_df[["Country name", "Year", "Population"]]
population_only_df = pd.DataFrame(selected_columns)
population_only_df.head()
```

```python
# Remove all countries except for one
country_name = "Rwanda"
one_country_df = population_only_df[population_only_df["Country name"] == country_name]
one_country_df.head()
```

```python
# Slice the DataFrame to include only rows where 'Year' is within the range of 1989 to 2021
one_country_df = one_country_df[(one_country_df['Year'] >= 1989) & (one_country_df['Year'] <= 2021)]
one_country_df
```

```python
# Reset Index
one_country_df.reset_index(drop=True, inplace=True)
one_country_df.head()
```

```python
# Drop Country Name and rename df
one_country_df.drop(["Country name"], axis=1, inplace=True)
rwanda_df = one_country_df
rwanda_df
```

```python
def millions(x, pos):
    return f"{x*1e-6:1.0f}M"
```

```python
# Create graph for lines to go on
fig, ax1 = plt.subplots(figsize=(10, 6))
```

```python
# Plot Population
ax1.plot(rwanda_df.Year, rwanda_df.Population, label="Population", color="blue")
```

```python
# Format to fit data
ax1.set_ylim(5000000, 14000000)
ax1.yaxis.set_major_formatter(millions)
```

```python
# Clean up Graph
ax1.set_xlabel("Year")
ax1.set_ylabel('People')
plt.ylabel('People')
plt.title("Rwanda: Overall Population 1989-2021")
ax1.grid(True)
```
